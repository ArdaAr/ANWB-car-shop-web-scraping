```python
BASE_URL = 'https://www.anwb.nl'

response = requests.get(f"{BASE_URL}/v2/privatelease")
print(response.json())
```

```python
manufactures = []
models = []
prices = []
fuels = []
chassis = []
conditions = []

for item in response.json()['items']:
    manufactures.append(item['manufacturer'])
    
    models.append(item['model'])
    
    fuels.append(item['fuelType'])
    
    prices.append(item['price'])
    
    chassis.append(item['vehicleChassis'])
    
    conditions.append(item['condition'])
```

```python
df = pd.DataFrame({
    'Manufacture' : manufactures,
    'Model' : models,
    'Price' : prices,
    'Benzine' : fuels,
    'Chassis' : chassis,
    "Status" : conditions
})
df
```
